```python
from fastai.vision import *
```

```python
folder = 'a320'
file = 'a320.txt'
```

```python
folder = 'b737'
file = 'b737.txt'
```

```python
path = Path('fighters')
dest = path/folder
dest.mkdir(parents=True, exist_ok=True)
```

```python
path.ls()
```

```python
classes = ['a320', 'b737']
```

```python
#download_images(path/file, dest, max_pics=200)
```

```python
for c in classes:
  print(c)
  verify_images(path/c, delete=True, max_size=500)
```

```python
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.2,
        ds_tfms=get_transforms(), size=224, num_workers=2).normalize(imagenet_stats)
```

```python
data.classes
```

```python
data.show_batch(rows=3, figsize=(7,8))
```

```python
data.classes, data.c, len(data.train_ds), len(data.valid_ds)
```

```python
learn = cnn_learner(data, models.resnet101, metrics=error_rate)
```

```python
learn.fit_one_cycle(50)
```

```python
learn.save('stage-1')
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(10, max_lr=slice(1e-05,1e-04))
```

```python
learn.save('stage-2')
```

```python
learn.load('stage-2')
```

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_confusion_matrix()
```

```python
interp.plot_top_losses(9, figsize=(30,22))
```
